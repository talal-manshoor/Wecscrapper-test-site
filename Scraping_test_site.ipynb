```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
link = "https://www.webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
r = requests.get(link)
print(r)
```

```python
soup = BeautifulSoup(r.text,"lxml")
print(soup)
```

```python
dives = soup.find_all("div",class_ = "col-sm-4 col-lg-4 col-md-4")
print(dives)
```

```python
print(len(dives))
```

```python
p_name = soup.find_all("a", class_ = "title")
ind = 0
for i in p_name:
  p_name[ind] = i.text.strip("...")
  print(p_name[ind])
  ind = ind + 1
```

```python
p_price = soup.find_all("h4", class_ = "pull-right price")
ind = 0
for i in p_price:
  p_price[ind] = i.text
  print(p_price[ind])
  ind = ind+1
```

```python
p_descrip = soup.find_all("p", class_ = "description")
ind = 0
for i in p_descrip:
    p_descrip[ind] = i.text
    print(p_descrip[ind])
    ind = ind + 1
```

```python
data = {'Title' : p_name, 'Price' : p_price, 'Description' : p_descrip }
df = pd.DataFrame(data)
print(df.head())
df.to_csv('Scraped_Data.csv',index=False)
```
